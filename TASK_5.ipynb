```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "acc_16.csv"  # Update with actual path
df = pd.read_csv(file_path, encoding="utf-8")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['DAY_WEEK'], palette='viridis')
plt.title("Accident Distribution by Day of the Week")
plt.xlabel("Day of the Week (1=Sunday, 7=Saturday)")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(x=df['WEATHER1'], palette='magma', order=df['WEATHER1'].value_counts().index)
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(x=df['REL_ROAD'], palette='cividis', order=df['REL_ROAD'].value_counts().index)
plt.title("Accidents by Road Condition")
plt.xlabel("Road Condition")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```
